```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

```python
def f(x, y):
    return np.cos(x)*np.sin(y)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 50)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contourf(X, Y, Z, 20, cmap='BrBG')
plt.colorbar();
plt.title("Ejemplo de contour map (cos(x) * sen(y))")
plt.xlabel("Valores de 0 a 5")
plt.ylabel("Valores de 0 a 5")
plt.savefig('contour_map.png')
```
